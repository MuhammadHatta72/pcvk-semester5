```python
# import data
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Fungsi dilasi tanpa library morphology dari OpenCV
def dilasi(citra, sed):
    tinggi, lebar = citra.shape
    setengah_tinggi_sed, setengah_lebar_sed = sed.shape[0] // 2, sed.shape[1] // 2
    hasil_dilasi = np.zeros((tinggi, lebar), dtype=np.uint8)

    for y in range(setengah_tinggi_sed, tinggi - setengah_tinggi_sed):
        for x in range(setengah_lebar_sed, lebar - setengah_lebar_sed):
            bagian_citra = citra[y - setengah_tinggi_sed:y + setengah_tinggi_sed + 1,
                                x - setengah_lebar_sed:x + setengah_lebar_sed + 1]
            hasil_dilasi[y, x] = np.max(bagian_citra * sed)

    return hasil_dilasi

# Baca citra plat nomor
citra = cv2.imread('/content/drive/MyDrive/PCVK/Meet9/plat nomer.jpg', cv2.IMREAD_GRAYSCALE)

# Buat Structuring Element (SED) berbentuk square 3x3
sed = np.ones((3, 3), dtype=np.uint8)

# Lakukan dilasi pada citra
hasil_dilasi = dilasi(citra, sed)

# Tampilkan citra asli dan hasil dilasi
plt.subplot(1, 2, 1)
plt.imshow(citra, cmap='gray')
plt.title('Citra Asli')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(hasil_dilasi, cmap='gray')
plt.title('Hasil Dilasi')
plt.axis('off')

plt.show()
```

```python
# Operasi opening dengan Menggunakan Library Morphology OpenCV
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca citra
citra = cv2.imread('/content/drive/MyDrive/PCVK/Meet9/plat nomer.jpg', cv2.IMREAD_GRAYSCALE)

# Konversi citra ke citra biner menggunakan thresholding
threshold_value = 128  # Nilai ambang threshold, sesuaikan sesuai kebutuhan
_, citra_biner = cv2.threshold(citra, threshold_value, 255, cv2.THRESH_BINARY)

# Buat Structuring Element (SED) berbentuk square 3x3
sed = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))

# Lakukan operasi opening dengan OpenCV pada citra biner
hasil_opening_opencv = cv2.morphologyEx(citra_biner, cv2.MORPH_OPEN, sed)

# Tampilkan citra asli, citra biner, dan hasil opening
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(citra, cmap='gray')
plt.title('Citra Asli')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(citra_biner, cmap='gray')
plt.title('Citra Biner')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(hasil_opening_opencv, cmap='gray')
plt.title('Hasil Opening (OpenCV)')
plt.axis('off')

plt.show()
```

```python
# Skeleton
flower = cv2.imread("/content/drive/MyDrive/PCVK/Meet9/lily.jpg", 0)
size = np.size(flower)
skel = np.zeros(flower.shape,np.uint8)

ret,img = cv2.threshold(flower,127,255,0)
element = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
done = False

while( not done):
    eroded = cv2.erode(img,element)
    temp = cv2.dilate(eroded,element)
    temp = cv2.subtract(img,temp)
    skel = cv2.bitwise_or(skel,temp)
    img = eroded.copy()

    zeros = size - cv2.countNonZero(img)
    if zeros==size:
        done = True
# Display the original and skeletonized images
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.imshow(flower, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(122)
plt.imshow(skel, cmap='gray')
plt.title("Skeletonized Image")
plt.axis("off")

plt.show()
```
