```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
path = '/content/drive/MyDrive/Kuliah/Semester 5/PCVK/Pratikum 2/ktp-random.jpg'
img = cv.imread (path)
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img)

# tutup bagian foto
cv.rectangle(img,pt1=(550,80), pt2=(770,330), color=(255,255,255), thickness=-1)
plt.imshow(img)

# tutup bagian agama da status perkawinan
cv.rectangle(img,pt1=(190,280), pt2=(500,320), color=(255,255,255), thickness=-1)
plt.imshow(img)
```
